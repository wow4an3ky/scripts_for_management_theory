Рекуррентный метод наименьших квадратов с линейной аппроксимацией:


```python
import numpy as np

n = int(input('Введите размерность (совпадает с количеством измерений) задачи: '))

x = np.zeros((2), dtype=float)
print('Ввод первых двух входов')
for i in range(2):
    x[i] = float(input(f'Элемент x{i+1}: '))

y = np.zeros((2), dtype=float)
print('Ввод первых двух выходов')
for i in range(2):
    y[i] = float(input(f'Элемент y{i+1}: '))

C_0 = np.vstack([x, np.ones(2)]).T
coeff = []
coeff.append(np.linalg.inv(C_0.T @ C_0) @ C_0.T @ y)
print(f'a_0 = {coeff[0][0]}, b_0 = {coeff[0][1]}')

H = C_0.T @ C_0

for i in range(2, n):
    #Новые измерения добавляем
    x_el = float(input(f'Измерение со входа {i+1}: '))
    y_el = float(input(f'Измерение с выхода {i+1}: '))
    C = np.array([[x_el, 1]])
    H = H + C.T @ C
    new_coeff = coeff[-1] + np.linalg.inv(H) @ C.T @ (y_el - C @ coeff[-1])
    coeff.append(new_coeff)
    print(f'a_{i-1} = {new_coeff[0]}, b_{i-1} = {new_coeff[1]}')
```

Рекуррентный метод наименьших квадратов с квадратичной аппроксимацией:

```python
import numpy as np

n = int(input('Введите размерность (совпадает с количеством измерений) задачи: '))

x = np.zeros((2), dtype=float)
print('Ввод первых двух входов')
for i in range(2):
    x[i] = float(input(f'Элемент x{i+1}: '))

y = np.zeros((2), dtype=float)
print('Ввод первых двух выходов')
for i in range(2):
    y[i] = float(input(f'Элемент y{i+1}: '))

C_0 = np.vstack([x**2, x, np.ones(2)]).T
coeff = []
coeff.append(np.linalg.inv(C_0.T @ C_0) @ C_0.T @ y)
print(f'a_0 = {coeff[0][0]}, b_0 = {coeff[0][1]}, c_0 = {coeff[0][2]}')

H = C_0.T @ C_0

for i in range(2, n):
    #Новые измерения добавляем
    x_el = float(input(f'Измерение со входа {i+1}: '))
    y_el = float(input(f'Измерение с выхода {i+1}: '))
    C = np.array([[x_el**2, x_el, 1]])
    H = H + C.T @ C
    new_coeff = coeff[-1] + np.linalg.inv(H) @ C.T @ (y_el - C @ coeff[-1])
    coeff.append(new_coeff)
    print(f'a_{i-1} = {new_coeff[0]}, b_{i-1} = {new_coeff[1]}, c_{i-1} = {new_coeff[2]}')
```
