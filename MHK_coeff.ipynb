Метод наименьших квадратов, который ищет оценку коэффоциенты для линейной и квадратичной аппроксимации:


```python
import numpy as np

n = int(input('Введите размерность (совпадает с количеством измерений) задачи: '))

#Ввод измерений входа X
x = np.zeros((n), dtype=float)
print("Ввод измерений входа x: ")
for i in range(n):
    x[i] = float(input(f"Элемент x{i+1}: "))
print(f"Столбец x выглядит так:\n{x}")

#Ввод измерений выхода Y
y = np.zeros((n), dtype=float)
print("Ввод измерений выхода y: ")
for i in range(n):
    y[i] = float(input(f"Элемент y{i+1}: "))
print(f"Столбец y выглядит так:\n{y}")

#Если линейная аппроксимация
C = np.vstack([x, np.ones(n)]).T
coeffs = np.linalg.lstsq(C, y, rcond=None)[0]
a, b = coeffs
print("Линейная аппроксимация:")
print(f"a = {a}, b = {b}")

#Квадратичная аппроксимация
A = np.vstack([x**2, x, np.ones(n)]).T
coeffs = np.linalg.lstsq(A, y, rcond=None)[0]
a, b, c = coeffs
print("Квадратичная аппроксимация:")
print(f"a = {a}, b = {b}, c = {c}")
```
