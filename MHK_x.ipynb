Метод наименьших квадратов, который ищет оценку x_MHK:


```python
import numpy as np

n = int(input('Введите размерность задачи: '))
k = int(input('Введите количество измерений: '))

#Запись измерений
y = {}
for i in range(k):
    yk = []
    for j in range(n):
        yk_inp = float(input(f'y{i+1}_{j+1} = ')) #i - номер измерения, j - позиция в векторе
        yk.append(yk_inp)
    print(f'y{i+1} = {yk}')
    y[f'y{i+1}'] = yk

#Стакаем все измерения в один столбец
Y = []
for key, value in y.items():
    Y.append(value)
Y = np.hstack(Y)

#Матрица C для одного измерения
C_block = np.zeros((n, n), dtype=float)
print("Ввод матрицы C_block")
for i in range(n):
    for j in range(n):
        C_block[i][j] = input(f"Элемент C_block{i+1, j+1}: ")
print(f"Матрица C_block выглядит так:\n{C_block}")

#Стакаем матрицы в общую (блочно)
C = np.vstack([C_block] * k)

# МНК решение: x = (C^T C)^(-1) C^T Y
x_hat = np.linalg.inv(C.T @ C) @ C.T @ Y

print("Оценка параметров x:")
print(x_hat)
```
